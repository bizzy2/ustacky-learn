```python
NigCovid = 'https://covid19.ncdc.gov.ng/'
page = requests.get(NigCovid)
print(page.text)
```

```python
soup = BeautifulSoup(page.content, "html.parser")
print(soup.prettify())
```

```python
result = soup.find_all('div', class_ ="row")
print(result)
```

```python
for row in result:
    rows = row.find_all('td')
    rows = np.array(rows)
    print(rows)
```
